# Unidade 1 - Introdução e Conceitos Básicos de Controle de Ruído

Para uma onda quadrada, uma onda senoidal e
uma onda triangular, gere gráficos das formas de onda. Considere a amplitude das ondas como
sendo unitária. Depois calcule o valor de RMS de cada uma delas, usando os valores dos vetores
das formas de onda (amostras). Compare os resultados obtidos com valores analíticos esperados
para o RMS. Não utilizar funções prontas para resolver os exercícios (com exceção da função
seno).

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
import IPython.display as ipd
```

```python
Tp = 1  # Tempo total da análise (s)
freq = 2  # Frequencia fundamental
fs = 44100 # Frequencia amostragem
t = np.linspace(0, Tp, int(fs*Tp)) # vetor temporal
```

```python
# Parâmetros dos sinais
A = 1.0  # Amplitude do sinal senoidal

# Sinal x2(t) - onda quadrada
x2_t = 0
for i in range(1, 1000):
        x2_t = x2_t + np.sin(2*np.pi*freq*t*(2*i-1))/((2*i-1))
x2_t = x2_t*4/np.pi

# Sinal x1(t) = A * sin(wt)
x1_t = A*np.sin(2*np.pi*freq*t)

# Sinal x3(t) - onda triangular
x3_t = 0
for i in range(1, 1000):
        x3_t = x3_t + np.cos(2*np.pi*freq*(2*i-1)*t)/((2*i-1)**2)
x3_t = -2*x3_t*4/(np.pi**2)

# Cálculo do valor médio e RMS
x0_1 = np.mean(x1_t)
x1_rms = np.sqrt(np.mean(x1_t**2))

x0_2 = np.mean(x2_t)
x2_rms = np.sqrt(np.mean(x2_t**2))

x0_3 = np.mean(x3_t)
x3_rms = np.sqrt(np.mean(x3_t**2))
```

```python
# Plot dos sinais
plt.figure(figsize=(12, 6))

plt.subplot(2, 1, 1)
plt.plot(t, x2_t, label='onda quadrada')
plt.ylim((-1.1, 1.1))
plt.xlabel('Tempo [s]')
plt.ylabel('Amplitude')
plt.title('Sinal')
plt.legend()

plt.subplot(2, 1, 2)
plt.plot(t, x2_t, label = r'$x(t)$ [-]')
plt.plot(t, x0_2*np.ones(len(t)), '--k', linewidth = 3, label = r'$x_0$ [-]')
plt.plot(t, x2_rms*np.ones(len(t)), '--r', linewidth = 3, label = r'$x_{rms}$ [-]')
plt.legend(loc = 'upper right')
plt.grid(linestyle = '--', which='both')
plt.ylim((-1.2*A, 1.2*A))
plt.xlim((t[0], t[-1]))
plt.ylabel('Valor')
plt.title('Valor Médio e RMS')
# plt.savefig('quadrada.pdf')
plt.tight_layout()

plt.show()
```

```python
print(f"Valor Médio de x2: {x0_2:.2f}")
print(f"RMS de x2: {x2_rms:.2f}")
```

```python
# Plot dos sinais
plt.figure(figsize=(12, 6))

plt.subplot(2, 1, 1)
plt.plot(t, x1_t, label='x(t) = A * sin(wt)')

plt.xlabel('Tempo [s]')
plt.ylabel('Amplitude')
plt.title('Sinal')
plt.legend()

plt.subplot(2, 1, 2)
plt.plot(t, x1_t, label = r'$x(t)$ [-]')
plt.plot(t, x0_1*np.ones(len(t)), '--k', linewidth = 3, label = r'$x_0$ [-]')
plt.plot(t, x1_rms*np.ones(len(t)), '--r', linewidth = 3, label = r'$x_{rms}$ [-]')
plt.legend(loc = 'upper right')
plt.grid(linestyle = '--', which='both')
plt.ylim((-1.2*A, 1.2*A))
plt.xlim((t[0], t[-1]))
plt.ylabel('Valor')
plt.title('Valor Médio e RMS')
# plt.savefig('seno.pdf')
plt.tight_layout()

plt.show()
```

```python
print(f"Valor Médio de x1: {x0_1:.2f}")
print(f"RMS de x1: {x1_rms:.2f}")
```

```python
# Plot dos sinais
plt.figure(figsize=(12, 6))

plt.subplot(2, 1, 1)
plt.plot(t, x3_t, label='onda triangular')
plt.ylim((-1.1, 1.1))
plt.xlabel('Tempo [s]')
plt.ylabel('Amplitude')
plt.title('Sinal')
plt.legend()

plt.subplot(2, 1, 2)
plt.plot(t, x3_t, label = r'$x(t)$ [-]')
plt.plot(t, x0_3*np.ones(len(t)), '--k', linewidth = 3, label = r'$x_0$ [-]')
plt.plot(t, x3_rms*np.ones(len(t)), '--r', linewidth = 3, label = r'$x_{rms}$ [-]')
plt.legend(loc = 'upper right')
plt.grid(linestyle = '--', which='both')
plt.ylim((-1.2*A, 1.2*A))
plt.xlim((t[0], t[-1]))
plt.ylabel('Valor')
plt.title('Valor Médio e RMS')
# plt.savefig('triangulo.pdf')
plt.tight_layout()

plt.show()
```

```python
print(f"Valor Médio de x3: {x0_3:.2f}")
print(f"RMS de x3: {x3_rms:.2f}")
```
